```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn

pd.set_option("display.width", 155)
```

```python
data_sgn = pd.read_csv("../logs/csv/SGN_20221128_165159/1.csv")
data_pp = pd.read_csv("../logs/csv/0PP_20221217_115518/1.csv")
print(data_sgn.head())
print(data_sgn.shape)
print(data_sgn.info())
```

```python
print(data_sgn.describe())
print("\n\nKurtosis:")
print(data_sgn.kurtosis(numeric_only=True))
```

```python
plt.figure(figsize=(16,4))
# grafica la densidad de los valores (distribución normal)
seaborn.kdeplot(x="ACCX", data=data_sgn)
plt.show()
plt.figure(figsize=(16,4))
seaborn.kdeplot(x="ACCX", data=data_pp)
plt.show()
```

```python
for i in data_sgn.columns:
    plt.figure(figsize=(15,6))
    # grafica la densidad de los valores (distribución normal)
    seaborn.kdeplot(x=i, data=data_sgn)
    seaborn.kdeplot(x=i, data=data_pp)
    plt.show()
```

```python
plt.figure(figsize=(20,6))
plt.plot([i for i in range(len(data_sgn["GYRZ"].dropna()))], data_sgn["GYRZ"].dropna())
plt.title("GYRZ")
plt.show()
```

```python
for d in data_sgn.columns:
    plt.figure(figsize=(8,6))
    plt.plot([i for i in range(len(data_sgn[d].dropna()))], data_sgn[d].dropna())
    #plt.xlabel("Tiempo (s*10e-2)")
    #plt.ylabel("m/s²")
    plt.title(d)
    plt.show()
```
