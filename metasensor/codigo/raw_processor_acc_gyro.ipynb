```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn

pd.set_option("display.width", 155)
```

```python
data = pd.read_csv("../logs/csv/sensor_SGN_2022-11-28 16:51:59.429743.csv")
print(data.head())
print(data.shape)
print(data.info())
```

```python
print(data.describe())
print("\n\nKurtosis:")
print(data.kurtosis(numeric_only=True))
```

```python
plt.figure(figsize=(15,6))
# grafica la densidad de los valores (distribución normal)
seaborn.kdeplot(x="GYRY", data=data)
plt.show()
```

```python
for i in data.columns:
    plt.figure(figsize=(15,6))
    # grafica la densidad de los valores (distribución normal)
    seaborn.kdeplot(x=i, data=data)
    plt.show()
```

```python
plt.figure(figsize=(20,6))
plt.plot([i for i in range(len(data["GYRZ"].dropna()))], data["GYRZ"].dropna())
plt.title("GYRZ")
plt.show()
```

```python
for d in data.columns:
    plt.figure(figsize=(8,6))
    plt.plot([i for i in range(len(data[d].dropna()))], data[d].dropna())
    #plt.xlabel("Tiempo (s*10e-2)")
    #plt.ylabel("m/s²")
    plt.title(d)
    plt.show()
```
