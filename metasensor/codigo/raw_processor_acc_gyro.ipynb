```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
```

```python
data = pd.read_csv("acc_gyro.log")
#data = pd.read_csv("acc_gyro.csv")
print(data.head())
print(data.shape)
print(data.info())
```

```python
print(data.describe())
#data.plot(kind="box", vert=False, figsize=(14,6))
```

```python
for i in data.columns:
    plt.figure(figsize=(8,6))
    # grafica la densidad de los valores (distribución normal)
    seaborn.kdeplot(x=i, data=data)
    plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.plot([i for i in range(len(data["ACCZ"]))], data["ACCZ"])
plt.xlabel("Tiempo (s*10e-2)")
plt.ylabel("m/s²")
plt.title("Acelerómetro - Eje Z")
plt.show()
```
