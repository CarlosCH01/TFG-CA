```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
```

```python
data = pd.read_csv("../logs/sensor_log.csv")
print(data.head())
print(data.shape)
print(data.info())
```

```python
print(data.describe())
#data.plot(kind="box", vert=False, figsize=(14,6))
```

```python
for i in data.columns:
    plt.figure(figsize=(8,6))
    # grafica la densidad de los valores (distribución normal)
    seaborn.kdeplot(x=i, data=data)
    plt.show()
```

```python
for d in data.columns:
    plt.figure(figsize=(8,6))
    plt.plot([i for i in range(len(data[d].dropna()))], data[d].dropna())
    #plt.xlabel("Tiempo (s*10e-2)")
    #plt.ylabel("m/s²")
    plt.title(d)
    plt.show()
```
