```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import neighbors, metrics
from sklearn.model_selection import train_test_split

import constants as CTS
```

```python
csv_data = pd.read_csv(CTS.CSV_DIR + "SGN_20221128_165159.csv", index_col=0)

X = csv_data[["ACCX","ACCY","ACCZ","GYRX","GYRY","GYRZ","MAGX","MAGY","MAGZ",]]
y = csv_data[["UID"]]
csv_data.isnull().sum()
```

```python
plt.figure(figsize=(15,6))
plt.plot(csv_data["MAGX"].dropna())
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=0)
```

```python
knn = neighbors.KNeighborsClassifier()
knn.fit(X_train,y_train)
prediction = knn.predict(X_test)
```

```python
metrics.accuracy_score(y_true = y_test, y_pred = prediction)
```
